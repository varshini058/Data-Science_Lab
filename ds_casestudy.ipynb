```python
file_path="C:\\Users\\user\\Downloads\\payroll_dataset.xlsx"
employees=pd.read_excel(file_path, sheet_name="Employees")
departments=pd.read_excel(file_path, sheet_name="Departments")
merged_data=pd.merge(employees,departments,on="DepartmentID")
merged_data.head()
```

```python
cleaned_data=merged_data.drop_duplicates()
cleaned_data
```

```python
min_sal=cleaned_data['Salary'].min()
max_sal=cleaned_data['Salary'].max()

cleaned_data['Normalized_sal']=(cleaned_data['Salary']-min_sal)/(max_sal-min_sal)
cleaned_data
```

```python
Q1=merged_data['Salary'].quantile(0.25)
Q3=merged_data['Salary'].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
IQR_data=merged_data[(merged_data['Salary']>=lower)&(merged_data['Salary']<=upper)]
IQR_data
```

```python
onehotencoded_data=pd.get_dummies(cleaned_data,columns=['Location'])
onehotencoded_data
```
