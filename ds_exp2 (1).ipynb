```python
import pandas as pd
data=pd.read_csv('housing_dataset.csv')
data
```

```python
data.isna().sum()
```

```python
median_val=data['Price'].median()
median_val
```

```python
data['Price']=data['Price'].fillna(median_val)
data
```

```python
mean_val=data['Bathrooms'].mean()
mean_val
```

```python
data['Bathrooms']=data['Bathrooms'].fillna(mean_val)
data
```

```python
data.head(13)
```

```python
data=data.dropna(axis=0)
data
```

```python
Q1=data['Price'].quantile(0.25)
Q3=data['Price'].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
outlier=((data['Price']<lower)|(data['Price']>upper))
```

```python
outlier
```

```python
cleaned_data=data[~outlier]
```

```python
cleaned_data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
cleaned_data['normalised']=scaler.fit_transform(cleaned_data[['Sqft_Living']])
cleaned_data
```

```python
cleaned_data
```

```python
cleaned_data['Year_Built']=cleaned_data['Year_Built'].astype(int)
cleaned_data
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
cleaned_data['Waterfront_encoded']=label_encoder.fit_transform(cleaned_data['Waterfront'])
cleaned_data
```

```python
cleaned_data.to_csv('housing_data_cleaned.csv',index=False)
```
